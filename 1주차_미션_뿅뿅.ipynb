Q5. ë‹¨ì–´ë“¤ì„ ì…ë ¥ë°›ì•˜ëŠ”ë°, ìê¾¸ë§Œ ìˆ«ìë“¤ì´ ì„ì—¬ë“¤ì–´ê°€ëŠ” ë¬¸ì œê°€ ìˆìŠµë‹ˆë‹¤. ì´ë¥¼ ì²˜ë¦¬í•˜ê¸° ìœ„í•´ì„œ í•¨ìˆ˜ì— stringì„ ì „ë‹¬ ë°›ì€ ë’¤, string ì•ˆì—ì„œì˜ ìˆ«ìë¥¼ ì œê±°í•œ í›„ stringë§Œ ë‚¨ì€ ë¦¬ìŠ¤íŠ¸ë¥¼ ì¶œë ¥í•˜ì„¸ìš”. ğŸ˜

- string ì—°ì‚°ì„ ì´ìš©í•´ì„œ ë¬¸ìì—´ì„ ìë¥´ëŠ” ì—°ì‚°ì„ ì‚¬ìš©í•´ë³´ì„¸ìš”.

```python
inputs = "cat32dog16cow5"
```

```python
def find_string(inputs):
  answer = []

  for string in inputs:
    if string.isalpha():
      answer.append(string)
    elif string.isdigit():
      answer.append(',')

  answer = ''.join(answer)
  answer = list(answer.split(','))

  remove = ['']
  answer = [i for i in answer if i not in remove]

  return answer
```

```python
find_string(inputs)
```
